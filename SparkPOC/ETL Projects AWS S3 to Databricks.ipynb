```
from urllib.parse import urlparse
import os
import boto3


client = boto3.client('s3'); 

def upload_to_s3(sourceFile_path, target_s3_url):
  
  
  
  
  #extracting bucket_name from s3 url.
  parsed_url = urlparse(target_s3_url, allow_fragments = 'false');  
  bucket_name = parsed_url.netloc;
  
  if(len(parsed_url.path)>0):
    target_path = (parsed_url.path + '/' + os.path.split(sourceFile_path)[1])[1:];
  else:
    target_path = os.path.split(sourceFile_path)[1];
  

  
  #uploading to s3:
  client.upload_file(sourceFile_path, bucket_name, target_path) #if file already exists in s3, it will replace.
  print("Uploaded to " + bucket_name + '/' + target_path )
```

```
upload_to_s3('dbfs:/user/hive/warehouse/sparkpoc.db/egptytousflightfile', 's3://my-spark-poc-learning/')
```
